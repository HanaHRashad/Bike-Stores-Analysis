### 4] Data Analysis & Visualization (EDA & Statistical Analysis)
#### Exploratory Data Analysis

##### What is the average Revenue?

```python
df['revenue'].mean()
```

##### What were the highest and lowest Revenue?

```python
#Highest
df['revenue'].max()
```

```python
#Lowest
df['revenue'].min()
```

##### How many orders have 'Road Bikes'  as a category name ?

```python
df[df['category_name']=='Road Bikes']['order_id'].count()
```

##### Display the data which revenue greater than 10000

```python
df[df['revenue']>10000].head(10)
```

##### what is the total revenue for each Brand ?

```python
df.groupby('brand_name')['revenue'].sum().sort_values(ascending=False)
```

##### what is the total revenue for each Category Name ?

```python
df.groupby('category_name')['revenue'].sum().sort_values(ascending=False)
```

Number of sales representatives

```python
df['sales_rep'].value_counts()
```

Top 2 sales representatives by revenue

```python
df.groupby('sales_rep')['revenue'].sum().sort_values(ascending=False).head(2)
```

Top 2 brands by revenue

```python
df.groupby('brand_name')['revenue'].sum().sort_values(ascending=False).head(2)
```

Best Selling Products

```python
df.groupby('product_name')['revenue'].sum().sort_values(ascending=False).head(10)
```

Most Consuming Cities

```python
df.groupby('city')['revenue'].sum().sort_values(ascending=False).head(10)
```

Total revenue by states

```python
df.groupby('state')['revenue'].sum()
```
